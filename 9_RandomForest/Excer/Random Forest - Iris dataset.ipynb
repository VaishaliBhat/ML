```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
df=pd.DataFrame(iris.data)
df
```

```python
df['target']=iris.target
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(df.drop(['target'],axis=1),df.target,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=10)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(df.drop(['target'],axis=1),df.target,test_size=0.35)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=10,criterion='entropy')
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y_predict=model.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
cm
```

```python
sn.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
plt.figure(figsize=(5,7))
```

```python
from sklearn.ensemble import RandomForestClassifier
# model=RandomForestClassifier(n_estimators=10,criterion='gini')
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(df.drop(['target'],axis=1),df.target,test_size=0.25)
```

```python
score=[]
n_estimators=range(1,101)
for i in n_estimators:
    model=RandomForestClassifier(n_estimators=i)
    model.fit(x_train,y_train)
    score.append(model.score(x_test,y_test))
    print('n_estimator= {}, score= {}'.format(i,score[i-1]))
    #print('score= {}, n_estimator= {}'.format(score[i-1],i))
plt.plot(n_estimators,score)
plt.xlabel('n_estimators')
plt.ylabel('testing accuracy')
```
