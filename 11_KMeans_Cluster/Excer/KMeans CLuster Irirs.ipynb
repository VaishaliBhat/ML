```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(iris.data,columns=iris['feature_names'])
df
```

```python
df['target']=iris.target
df
```

```python
df=df.drop(['sepal length (cm)','sepal width (cm)','target'],axis=1)
df
```

```python
plt.scatter(df['petal length (cm)'],df['petal width (cm)'])
```

```python
km=KMeans(n_clusters=3)
km
```

```python
km.fit_predict(df[['petal length (cm)','petal width (cm)']])
```

```python
y_predict=km.fit_predict(df[['petal length (cm)','petal width (cm)']])
y_predict
```

```python
df['cluster']=y_predict
df
```

```python
plt.plot(df.cluster,y_predict)
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],label='df1',color="green")
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],label='df1',color="black")
plt.scatter(df3['petal length (cm)'],df3['petal width (cm)'],label='df1',color="red")
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.legend()
```

```python
k_rng=range(1,10)
sse=[]
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df)
    sse.append(km.inertia_)
```

```python
plt.xlabel("K")
plt.ylabel("Sum of squared error")
plt.plot(k_rng,sse)
```
