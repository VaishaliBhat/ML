```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model 
```

```python
df= pd.read_csv("homeprices.csv")
df.head()
```

```python
#plot the graph

%matplotlib inline
plt.xlabel("Price per sqft")
plt.ylabel("area in sqft")
plt.title("House price per sqft")
plt.scatter(df.area,df.price,color='red',marker="+")
#plt.show
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
135.78767123*3300+180616.43835616432
```

```python
#Plotting along with linear regression

%matplotlib inline 
plt.xlabel('area (sqft)',fontsize=15)
plt.ylabel('Price (sqft)',fontsize=15)
plt.title("House area price prediction",fontsize=20,color="green")
plt.scatter(df.area,df.price,marker="+",color="red")
```

```python
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```
